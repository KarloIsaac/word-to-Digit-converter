### Import such libraries

```python
import pandas as pd
import numpy as np
```

##### Import file

```python
df = pd.read_csv("categorical_df.csv")
df.sample(5)
```

```python
df['month'].unique()
```

###### Make dictionary for month

```python
month_dict = {'jan':1, 'feb':2, 'mar':3, 'apr':4, 'may':5, 'jun':6, 'jul':7, 'aug':8, 'sep':9, 'oct':10, 'nov':11, 'dec':12}
print(month_dict)
```

### Convert month into number with help of month_dict data

```python
df["month"].replace(month_dict, inplace=True) #this is much faster than looping; also, in only takes ones line! ;)
```

```python
df.head()
```

```python
df['week day'].unique()
```

#### Make Dictionary for week day

```python
week_dict = {'sun':1, 'mon':2, 'tue':3, 'wed':4, 'thu':5, 'fri':6, 'sat':7}
week_dict
```

### Convert week day into number with help of week_dict data

```python
df["week day"].replace(week_dict, inplace=True) #one liner, much more efficient.
```

```python
df.head()
```

```python
from word2number import w2n
```

### Convert word number into digit number using word2number module

```python
df['number'] = df['number'].map(lambda weekday: w2n.word_to_num(weekday))
```

```python
df.head()
```

## Convert $data Type$ into int

```python
df['month'] = df['month'].astype(int)
df['week day'] = df['week day'].astype(int)
df['number'] = df['number'].astype(int)
```

```python
df.info()
```

# Sort value

```python
df = df.sort_values(by=['month','week day', 'number'], ascending=[False,True,False])
df
```

# to again convert number into word make dict for each one

```python
month_num_dict = {1:'jan', 2:'feb', 3:'mar', 4:'apr', 5:'may', 6:'jun', 7:'jul', 8:'aug', 9:'sep', 10:'oct', 11:'nov', 12:'dec'}
week_num_dict = {1:'sun', 2:'mon', 3:'tue', 4:'wed', 5:'thu', 6:'fri', 7:'sat'}
print(month_num_dict, '\n\n', week_num_dict)
```

```python
# Converting month
df['month'].replace(month_num_dict, inplace=True)

# Converting Week
df['week day'].replace(week_num_dict, inplace=True)
```
