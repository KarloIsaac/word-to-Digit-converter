```python
import pandas as pd

month_list = ['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']
week_list = ['sun', 'mon', 'tue', 'wed', 'thu', 'fri', 'sat']
numbers = ['one', 'two', 'three', 'four', 'five']

df = pd.read_csv("categorical_df.csv")
df['month'] = df['month'].astype(pd.CategoricalDtype(categories=month_list, ordered=True))
df['week day'] = df['week day'].astype(pd.CategoricalDtype(categories=week_list, ordered=True))
df['number'] = df['number'].astype(pd.CategoricalDtype(categories=numbers, ordered=True))

df.sort_values(by=['month', 'week day', 'number'], ascending=[False, True, False], inplace=True)
df
```
